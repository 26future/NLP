```python
# 로그인 하기
driver = webdriver.Chrome("chromedriver")
driver.implicitly_wait(3)               # 드라이버가 초기화 될 때까지 3초 대기
driver.get('https://www.danawa.com/')   # 크롬 웹브라우저를 새로 연다

# 로그인 스크립트 소스를 찾아냄
login = driver.find_element_by_css_selector('li.my_page_service > a')
login.click()
driver.implicitly_wait(3)

my_id = "---본인 아이디 입력---"
my_pw = "---본인 패스워드 입력---"

# ID와 Password를 입력하여 클릭
# <input type="text" id="danawa-member-login-input-id" maxlength="40" name="id" class="int" placeholder="아이디/인증 이메일" value="" style="z-index: 8;">
driver.find_element_by_id("danawa-member-login-input-id").send_keys(my_id)
driver.implicitly_wait(2)

# <input type="password" maxlength="20" id="danawa-member-login-input-pwd" name="password" class="int" placeholder="비밀번호" style="z-index: 8;">
driver.find_element_by_name('password').send_keys(my_pw)
driver.implicitly_wait(2)

# <button class="btn_login" id="danawa-member-login-loginButton" onclick="return false;">로그인</button>
driver.find_element_by_css_selector('button.btn_login').click()
driver.implicitly_wait(2)

# 관심 상품 링크 클릭
# <li class="interest_goods_service">
# 			<a href="http://www.danawa.com/myPage/wishList.php" class="">
# 				<span class="my_serv_tit">관심상품</span>
# 				<span class="number_has">4</span>			</a>
# </li>
wishlist = driver.find_element_by_css_selector("li.interest_goods_service > a").click()
driver.implicitly_wait(2)

html_src = driver.page_source # 현재 페이지의 소스
print(html_src)

input()

driver.close()
```

### 한국은행 경제통계시스템 통계지표 활용하기 1

```python
from selenium import webdriver
import time

driver = webdriver.Chrome("chromedriver")
driver.implicitly_wait(3)              


# 한눈에 보는 우리나라 100대 통계지표
driver.get("http://ecos.bok.or.kr/jsp/vis/keystat/#/key") 

# <img src="../img/btn_excel_download.png" style="margin-top:-5px;" alt="download">
excel_download = driver.find_element_by_css_selector('img[alt="download"]')
driver.implicitly_wait(3)

excel_download.click()
print("파일 다운로드 실행 중....")
time.sleep(5)

driver.close()
```

### 한국은행 경제통계시스템 통계지표 활용하기 2

```python
from bs4 import BeautifulSoup
import time

   
item_found = 0 

while not item_found:

    keyword = ""       
    while len(keyword) == 0:
        keyword = str(input("검색할 항목을 입력하세요: "))  # 'CD' 를 입력한다.'기준금리'
    # keyword = "CD"

    driver = webdriver.Chrome("chromedriver")
    driver.implicitly_wait(3)
    driver.get("http://ecos.bok.or.kr/jsp/vis/keystat/#/key") 
    time.sleep(5)  
    
    # <a href="" data-ng-click="key.addTab(list.code,list.name,list.type,list.cate,list.cycle)" class="ng-binding">CD수익률(91일)<span data-ng-show="key.showTime" class="ng-binding"> ('20.08.28)</span></a>
    items1 = driver.find_elements_by_css_selector('a[class="ng-binding"]')
    
#     <a class="a-c1-list ng-binding" href="" data-ng-click="key.addTab(list.code,list.name,list.type,list.cate,list.cycle)">경제성장률(실질, 계절조정 전기대비)
# 										<span data-ng-show="key.showTime" class="ng-binding">('20.2/4)</span>
# 									</a>
#     items2 = driver.find_elements_by_css_selector('a[class="a-c1-list ng-binding"]')

# <a class="a-c1-list ng-binding" href="" data-ng-click="key.addTab(list.code,list.name,list.type,list.cate,list.cycle)">경제성장률(실질, 계절조정 전기대비)
# 										<span data-ng-show="key.showTime" class="ng-binding">('20.2/4)</span>
# 									</a>
#     items3 = driver.find_elements_by_css_selector('a[class="a-c4-list ng-binding"]')
    driver.implicitly_wait(3)
    
    print(len(items1))  # 75
    
    # input()
    #items = items1[1:] + items2 + items3    
    items = items1[1:]   
    #print(items[0].text)
    
    for idx, item in enumerate(items):
        # print(item.text)
        if keyword in item.text:
            print("검색어 '%s'에 매칭되는 '%s' 통계지표를 검색 중..." % (keyword, item.text))
            item.click()
            item_found = 1
            time.sleep(5)
            break
        elif idx == (len(items) - 1):
            print("검색어 '%s'에 대한 통계지표가 존재하지 않습니다..." % keyword)
            driver.close()
            continue
        else:
            pass


# input()

html_src = driver.page_source
soup = BeautifulSoup(html_src, 'html.parser')
driver.close()

# 'CD 수익률(91일)' 테이블
# <td class="ng-binding">2013</td>
table_items = soup.find_all('td', {'class':'ng-binding'})
# print(table_items)
date = [t.text for i, t in enumerate(table_items) if i % 3 == 0]    # table의 첫번째 항목 : '시점'
value = [t.text for i, t in enumerate(table_items) if i % 3 == 1]   # table의 두번째 항목 : '지표'  
change = [t.text for i, t in enumerate(table_items) if i % 3 == 2]  # table의 세번째 항목 : '전기대비증감' 

result_file = open('bok_statistics_%s.csv' % keyword, 'w') # bok : bank of Korea,한국은행

for i in range(len(date)):
    result_file.write("%s, %s, %s" % (date[i], value[i], change[i]))
    result_file.write('\n')

result_file.close()
print("키워드 '%s'에 대한 통계지표를 저장하였습니다." % keyword)

print(date)
print(value)
print(change)
```
